## Core Concepts:

*  **Sample space** contains list (set) of all possible outcomes $\Omega$
  * Mutually exhaustive
  * Collectively exhaustive
  * At the right granularity  
  
  
* **Event:** a subset of the sample space
  
  
* **Fundamental Axioms:**
  1. Nonnegativity: $P(A) >= 0$
  2. Normalization: $P(\Omega) = 1$
  3. Finite (Countable) Additivity: If $A \cap B = \emptyset$, then $P(A \cup B) = P(A) + P(B)$
    * Generally: $A \cap B \neq \emptyset$, then $P(A \cup B) = P(A) + P(B) - P(A \cap B)$
    * **NOTE:** additivity only holds for countable (i.e. sequencable) sequences of events (works for area, but not unit square, real line, etc.)
  
  
* **Consequences:**
  * $P(A) <= 1$
  * $P(\emptyset) = 0$
  * $P(A) + P(A^c) = 1$
  * $P(\{s_1,s_2,...,s_k\}) = P(s_1) + P(s_2) + ... + P(s_k)$
  * If $A \subset B$, then $P(A) <= P(B)$
  * $P(A \cup B \cup C) = P(A) + P(A^c \cap B) + P(A^c \cap B^c \cap C)$
  
  
* **Discrete Uniform Law:** if $\Omega$ consists of n equally likely elements, and event A consists of k elements. then: $\mathbf{P(A) = k * \frac{1}{n}}$
  
  
* For a continuous sample space, **Probability = Area** (Uniform Probability Law)
  
  
* **Probability calculations:**
  1. Specify sample space
  2. Specify probability law
  3. Identify event of interest
  4. Calculate outcome
  
  
* Probability can be interpreted as:
  * Frequencies
  * Descriptions of beliefs
  * Betting preferences

## Mathematical Background

### Sets:
* A collection of distinct elements
  * e.g. {a, b, c, d} (finite set); Real numbers (infinite set)
  

* $x \in S$ --> x is an element of S (and v.v., $x \notin S$)


* Specifying a set: {$x \in S: \cos{x} > \frac12$}


* Universal set: $\Omega$


* Complement of $S$: $S^c$
  * $x \in S^c$ if $x \in \Omega$ AND $x \notin S$
  * $(S^c)^c = S$


* Empty set: $\Omega^c = \emptyset$


* Subsets: S is a subset of T if $S \subset T$ (if $x \in S$, then $x \in T$)
  * S can be smaller or equal to T
  
  
* Unions: $x \in S \cup T$ implies $x \in S$ or $x \in T$


* Intersections: $x \in S \cap T$ implies $x \in S$ and $x \in T$


* Infinite collection of sets: $S_n,\ n = 1,2,3...$
  * $x \in \bigcup\limits_{n} S_n$ if  $x \in  S_n$ for some n 
  * $x \in \bigcap\limits_{n} S_n$ if  $x \in  S_n$ for all n
  
  
* Distributive set properties:
  * $S \cap (T \cup U) = (S \cap T) \cup (S \cap U)$
  * $S \cup (T \cap U) = (S \cup T) \cap (S \cup U)$
  
  
### De Morgan's Laws:
* Allow transfer between unions and intersections:


1. $(S \cap T)^c = S^c \cup T^c$
  * Or formally, $(\bigcap\limits_{n}S_n)^c = \bigcup\limits_{n}S_n^c$
  
  
2. $(S \cup T)^c = S^c \cap T^c$
  * Or formally, $(\bigcup\limits_{n}S_n)^c = \bigcap\limits_{n}S_n^c$
  
  
### Sequences and their Limits:
* Any kind of set possible, e.g. $S = \mathbb{R}$ (all real numbers), $S = \mathbb{R^n}$


* Sequence: a function that associates an element of a set S for any natural number i
  * Evaluate $f(i)$, get $a_i$ (i'th entry in sequence)
  

* Example sequences: $a_1, a_2, a_3,...$, or $a_i = \{a_i\}$ where $i \in \mathbb{N} = \{1, 2, 3,...\}$ and $a_i \in S$


* Typically care if sequence converges to certain number a: $\lim\limits_{i \to \infty}a_i = a$

## Lecture 1 Exercises:

**#1 Are these legitimate sample spaces?**

A. $\Omega$ = {Heads and rain, Heads and no rain, Tails}
* Mutually exclusive + collectively exhaustive, so yes

B. $\Omega$ = {Heads and rain, Tails and no rain, Tails}
* Not mutually exclusive --> No



**#2 Let A and B be events on the same sample space, with P(A)=0.6 and P(B)=0.7. Can these two events be disjoint?**
* No, as P(A) + P(B) > 1 (normalization axiom)


**#3 Let A, B, and C be disjoint subsets of the sample space. For each one of the following statements, determine whether it is true or false. Note: "False" means "not guaranteed to be true."**

A. $P(A) + P(A^c) + P(B) = P(A \cup A^c \cup B)$
* False

B. $P(A) + P(B) <= 1$
* True (disjoint, so finite additivity holds)

C. $P(A^c) + P(B) <= 1$
* False

D. $P(A \cup B \cup C) >= P(A \cup B)$
* True (disjoint, so finite additivity holds)


**#4 Let A, B, and C be subsets of the sample space, not necessarily disjoint. For each one of the following statements, determine whether it is true or false. Note: “False" means “not guaranteed to be true."**

A. $P((A \cap B) \cup (C \cap A^c)) <= P(A \cup B \cup C)$
* True --> left side is subset of right side

B. $P(A \cup B \cup C) = P(A \cap C^c) + P(C) + P(B \cap A^c \cap C^c)$
* True; see last consequence of axioms


**#5 Consider the same model of two rolls of a tetrahedral die, with all 16 outcomes equally likely. Find the probability of the following events:**

A. The value in the first roll is strictly larger than the value in the second roll.
* 3/8 = 0.375

B. The sum of the values obtained in the two rolls is an even number.
* 4/8 = 0.5


**#6 Consider a sample space that is the rectangular region [0,1]×[0,2], i.e., the set of all pairs (x,y) that satisfy 0 ≤ x ≤ 1 and 0 ≤ y ≤ 2. Consider a “uniform" probability law, under which the probability of an event is half of the area of the event. Find the probability of the following events:**

A. The two components x and y have the same values.
* P(line) = 0 (zero area)

B. The value, x, of the first component is larger than or equal to the value, y, of the second component.
* 0.25

C. The value of $x^2$ is larger than or equal to the value of y.
* Integral of $x^2$ from x = 0 to 1 = 1/3. Since total area is 2, P = 1/6.


**#7 Let the sample space be the set of positive integers and suppose that P(n)=1/2n, for n=1,2,…. Find the probability of the set {3,6,9,…}, that is, of the set of of positive integers that are multiples of 3.**
* Countable additivity: $\frac{1}{2^3} + \frac{1}{2^6} + \frac{1}{2^9} + ... = \alpha + \alpha^2 + \alpha^3 + ... = \frac{\alpha}{1-\alpha} = \frac{\frac18}{1 - \frac18} = \frac17$


**#8 Let the sample space be the set of positive integers. Is it possible to have a “uniform" probability law, that is, a probability law that assigns the same probability c to each positive integer?**
* No -- if c=0, $P(\Omega) = 1 = 0$. If c > 0, $P(\Omega) > 1$, which contradicts normalization.


**#9 Let the sample space be the two-dimensional plane. For any real number x, let $\mathbf{A_x}$ be the subset of the plane that consists of all points of the vertical line through the point (x,0), i.e., $\mathbf{A_x}$ = {(x,y) : y∈ℜ}.**

A. Do the axioms of probability theory imply that the probability of the union of the sets $A_x$ (which is the whole plane) is equal to the sum of the probabilities $P(A_x)$?
* No, because the real numbers can't be sequenced, therefore isn't countable and the additivity axiom does not hold

B. Do the axioms of probability theory imply that this is true when we consider only those lines for which the x coordinate is a positive integer?
* Yes, because 1 to $\infty$ is sequencable/countable, so additivity holds true
