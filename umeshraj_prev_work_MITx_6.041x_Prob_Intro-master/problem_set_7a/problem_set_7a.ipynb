## Problem 2.3a: Hypothesis test between two coins

```python
num_simu = 1000
num_correct = 0
for i_simu in range(num_simu):
    coin = np.random.choice(['c1', 'c2'], size=1, p=[2/3, 1/3])
    if coin == 'c1':
        toss = np.random.choice(['h', 't'], size=3, p=[1/3, 2/3])
    elif coin == 'c2':
        toss = np.random.choice(['h', 't'], size=3, p=[2/3, 1/3])
    
    num_heads = np.sum(toss == 'h')
    decision = 'c1' if num_heads <= 2 else 'c2'
    if decision == coin:
        num_correct += 1

print(num_correct/num_simu)
```

## Problem 5: Hypothesis test between two normals


```python
x = np.linspace(-5, 5)
sigma1 = 1
sigma2 = 2
g1 = 1/(np.sqrt(2*np.pi) * sigma1) * np.exp(-0.5*x**2/sigma1**2)
g2 = 1/(np.sqrt(2*np.pi) * sigma2) * np.exp(-0.5*x**2/sigma2**2)

plt.plot(x, g1)
plt.plot(x, g2)

# boundary
x_cross = np.sqrt(8/3 * np.log(2))
plt.vlines([-x_cross, x_cross], 0, 0.4)
```
