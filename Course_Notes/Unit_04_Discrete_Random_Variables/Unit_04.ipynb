## Core Concepts:

**Random Variable:** a numerical quantity whose value is determined by the outcome of the probabilistic experiment (e.g. the weight of a randomly selected person)  

* Abstract concept of weight: $\mathbf{W}$ (random variable)
  
* Particular weight of given student: $\mathbf{w}$ (actual number)  
  
* Weight is an object whose value is determined once the outcome of the prob. experiment is known  
  
* Can think of Weight ($W$) as black box/function: $student \to \mathbf{W(student)} \to w$  
  
* Possible for a given probabilistic experiment to have multiple associated random variables (e.g. weight W, height H, etc.)
  * Create new random variable based off other random variables, e.g. $BMI = B = \frac{W}{H^2}$ (still a function of outcomes)<br><br>
    
* Random variables are functions mapping $\Omega$ to real numbers (discrete or continuous values)  
  
    * Meaning of $X + Y$: random variable takes value $x + y$, where X takes value x and Y takes value y given some outcome<br><br>
      
* Since some of the outcomes of a probabilistic experiment are more likely than others, the associated numerical values of random variables will be more likely than others<br><br>
      
      
**Probability Mass Functions (PMF, Probability Law or Probability Distribution):** description of relative likelihoods for values of random variable  

* Define the y-axis of a PMF for random variable X, $p_X(x) = P(X=x) = P(\{\omega \in \Omega\ such\ that\ X(\omega)=x\})$  
  
  * $PMF = f(x)$, where for any x, the PMF specifies the probability that the random variable X takes on this particular value x<br><br>
    
* Properties:  
  
  * $p_X(x) \geq 0$ (like any probability)
    
  * $\sum\limits_{x}p_X(x)=1$

**Expectation:** one form of summary of a random variable (the average, mean, etc.) $\to$ the average outcome in a large number of independent trials
* E.g.: X is random variable of "gain" in a trial. $p_X(1) = \frac{2}{10}, \ \ p_X(2) = \frac{5}{10}, \ \ p_X(3) = \frac{3}{10}$
  * Can say the expected (average) gain is $\frac{1*200 + 2*500 + 3*300}{1000} = 1p_X(1) + 2p_X(2) + 3p_X(3)$
<br><br>
* Generally, $\mathbf{E[X] = \sum\limits_{x}xp_X(x)}$
* Assumption: for infinite sums, $E[X] = \sum\limits_{x}|x|p_X(x)<\infty$ (infinite sum well defined)
* Average of a population: e.g. n students, weight of i'th student is $x_i$
  * If all equally likely to be chosen and $x_i$'s are distinct: $E[X] = \sum\limits_{i}x_i\frac{1}{n} = \frac{1}{n}\sum\limits_{i}x_i$
  <br><br>
  
* Properties:
  * If for all $\omega, \ X(\omega) \geq 0$, that is, $X \geq 0$, then $E[X] \geq 0$
  * If $a \leq X \leq b$, then $a \leq E[X] \leq b$
  * For a constant c, $E[c] = c$
<br><br>


**Expectation of Bernoulli Random Variable:**
* Recall $p_X(1) = p$ and $p_X(0) = 1-p$
* As such, $E[x] = 1p + 0(1-p) = p \to \mathbf{E[X] = p}$
  * If $X$ is indicator variable $X = I_A$ (i.e. $X = 1$ if A occurs, $X=0$ otherwise, such that $p = P(A)$) $\to E[I_A] = P(A)$ 
<br><br>


**Expectation of Uniform Random Variable:**
  * E.g. uniform on [0,n] (a = 0, b = n): $p_X(x) = \frac{1}{n+1}$
    * $E[X] = \frac{1}{n+1}(0 + 1 + 2 + ... + n) = \frac{1}{n+1}\frac{n(n+1)}{2} = \frac{n}{2}$
      <br><br>
  * Generally, $E[X] = \frac{a+b}{2} \to$ the midpoint! AKA the "balance point"
<br><br>


**Expected Vale Rule for Calculating $\mathbf{E[g(X)]}$:**

```python
display(Image(filename='l4_5.png', embed=True, width=300))
```

  * Let $X$ be a random variable and $Y = g(X) \to$ 2 ways to calculate $E[Y]$:
    1. Averaging over y: $E[Y] = \sum\limits_{y}yp_Y(y)$
      * $E[Y] = 3*(0.1 + 0.2) + 4 * (0.3 + 0.4)$
      <br><br>
    2. Averaging over x (**Expected Value Rule**): $\mathbf{E[Y] = E[g(X)] = \sum\limits_{x}g(x)p_X(x)}$
    <br>
      * $E[Y] = 3*0.1 + 3*0.2 + 4*0.3 + *0.4$
      <br><br>
  * Number 2 has the advantage that we don't have to find the pmf of new variable ($p_Y(y)$ not needed)
  * Application:
  <br>
    * $E[X^2] = \sum\limits_{x}x^2p_X(x) \to$ here, $g(x) = x^2$
    * Note that $E[g(x)] \neq g(E[X])$, e.g. here $E[X^2] \neq (E[X])^2$
<br><br>


* **Linearity of Expectations:**
  * $E[aX + b] = aE[X]+b$
  * Can derive with Expected Value Rule
<br><br>


* **Variance:** quantifies the spread of the distribution of a random variable
  * "Distance from the mean $\mu$": Variance ~ $X - \mu$ (which is a random variable in itself)
    * Can calculate $E[X-\mu] = E[x] - \mu = 0$
    <br><br>
  * More mathematically sound $\to$ the squared distance from the mean: $\mathbf{var(X) = E[(X - \mu)^2]}$ $\ \ $(note: $var(X) \geq 0$)
  <br><br>
  * Calculate using Expected Value Rule $E[g(X)] = \sum\limits_{x}g(x)p_X(x)$ with $g(x) = (x - \mu)^2 \to \mathbf{var(X) = E[g(X)] = \sum\limits_{x}(x-\mu)^2p_X(x)}$
  * Properties:
    * Adding a constant to a random variable does not change the variance: $var(X + b) = var(X) \to$ just shifts PMF left or right
    * Multiplying random variable by constant does the following to variance: $var(aX) = a^2var(X)$
    <br><br>
  * Variance can be calculated just with means with handy formula: $\mathbf{var(X) = E[X^2] - (E[X])^2}$
    * Foil out $E[(X - \mu^2)]$ recalling $\mu = E[X]$ to see this is true
<br><br>


* **Standard Deviation:** more intuitive measure of spread with same units as random variable
  * Note that variance has units $[m^2]$ if random variable has units $[X]$.
    * Take square root of variance: $\sigma = \sqrt{var(X)}$
<br><br>


**Variance of Bernoulli Random Variable:**
* Recall $p_X(1) = p, \ p_X(0) = 1-p$, and $E[X] = p$ for Bernoulli random variable
* Two ways to calculate variance of Bernoulli:  
  
  1. $var(X) = \sum\limits_{x}(x-\mu)^2p_X(x)$
    * $var(X) = (1-p)^2p + (0 - p)^2(1-p) = p(1-p)$
  <br><br>
  2. $var(X) = E[X^2] - (E[X])^2$
    * Note that here, $X^2 = X$ (Bernoulli special case)
    * As such, $var(X) = E[X] - (E[X])^2 = p - p^2 = p(1-p)$
<br><br>    
* So $\mathbf{var(X) = p(1-p)}$ for Bernoulli -- zero variance at p = 0 and p = 1, max variance at p = 0.5
<br><br>


**Variance of Uniform Random Variable:**
* Use alternative formula (easiest math):
  * $var(X) = E[X^2] - (E[X])^2 = \frac{1}{n+1}(0^2 + 1^2 + 2^2+...+n^2) - (\frac{n}{2})^2$
  * Use formula for sum in that expression, get $var(X) = \frac{1}{12}n(n+2)$
    * Note that instead of 0 to n, we can say a to b (as long as n = b - a) and variance does not change! Just a PMF shift.  
    

* As such, make substitution: $\mathbf{var(X) = \frac{1}{12}(b-a)(b-a+2)}$
<br><br>


**Conditional PMF and Expectation:**
* Up to now:
  * $p_X(x) = P(X=x)$
  * $\sum\limits_{x}p_X(x) = 1$
  * $E[X] = \sum\limits_{x}xp_X(x)$
  * Expected Value Rule: $E[g(X)] = \sum\limits_{x}g(x)p_X(x)$
  * $var(X) = \sum\limits_{x}(x-E[X])^2p_X(x) = E[X^2] - (E[X])^2$
  <br><br>
* Conditioning on some event A (event A is known to have occured and $P(A) > 0$) 
  * $p_{X|A}(x) = P(X = x|A)$
  * $\sum\limits_{x}p_{X|A}(x) = 1$
  * $E[X|A] = \sum\limits_{x}xp_{X|A}(x)$
  * Expected Value Rule: $E[g(X)|A] = \sum\limits_{x}g(x)p_{X|A}(x)$
  * $var(X|A) = \sum\limits_{x}(x-E[X|A])^2p_{X|A}(x) = E[X^2|A] - (E[X|A])^2$
  <br><br>


**Total Expectation Theorem**:
* Recall Total Probability Theorem: $P(B) = P(A_1)P(B|A_1)+...+P(A_n)P(B|A_n)$
* If $B = \{X = x\}$:
  * $p_X(x) = P(A_1)p_{X|A_1}(x)+...+P(A_n)p_{X|A_n}(x) \to$ Total Probability Theorem translated to PMF notation
  <br><br>
* Multiplying both sides by $\sum\limits_{x}x$, get:
  * $\mathbf{E[X] = P(A_1)E[X|A_1]+...+P(A_n)E[X|A_n]}$
    * Allows dividing PMF into sections $A_1, A_2,...,A_n$ and "divide and conquer" to find $E[X]$
<br><br>


**Conditioning a Geometric Random Variable:**
* Recall Geometric Random Variable properties:
  * $P(H) = p$
  * Random variable $X$: number of trials until first heads

```python
display(Image(filename='l4_6.png', embed=True, width=400))
```

* "Memorylessness": number of remaining tosses, conditioned on T in first toss, is still geometric with parameter $p$
  * $p_{X-n|X>n}(k) = P(X-n=k|X>n) = P(T_2T_3T_4....H_{k+1}|T_1) = P(T_2T_3T_4....H_{k+1}) = p_X(k)$
  * Bottom line: $\mathbf{p_X(k) = p_{X-n|X>n}(k)}$, and as such, $\mathbf{E[X-n|X>n] = E[X]}$
  <br><br>
* Useful for calculating expectation of geometric distribution:
  * $E[X] = \sum\limits_{k=1}^\infty k p_X(k) = \sum\limits_{k=1}^\infty k (1-p)^{k-1}p$ is hard to compute!
  * Instead: $E[X] = 1 + E[X-1]$ (by linearity of expectations)
    * $E[X] = 1 + E[X-1] = 1 + pE[X-1|X=1] + (1-p)E[X-1|X>1]$ by Total Expectation Theorem
    * $E[X] = 1 + 0 + (1-p)E[X]$
    <br><br>
  * As such, $\mathbf{E[X] = \frac{1}{p}}$ for Geometric Random Variable X
<br><br>


**Multiple Discrete Random Variables and Joint PMFs:**
* Say we have $X:p_X$ and $Y:p_Y$, two individual random variables and their marginal (individual) PMFs
  * Cannot say anything about relationship between $X$ and $Y$ (e.g. $P(X = Y) = ?$)
  <br><br>
* Joint PMF: $p_{X,Y}(x,y) = P(X = x\ and\ Y = y)$ 
  * Function of 2 arguments $f(x,y)$
  * $\sum\limits_{x}\sum\limits_{y}p_{X,Y}(x,y) = 1$
  <br><br>


```python
display(Image(filename='l4_7.png', embed=True, width=300))
```

* Can deduce marginal (individual) PMFs from joint PMF: add probability rows/columns accordingly
  * $p_X(x) = \sum\limits_{y}p_{X,Y}(x,y)$ and $p_Y(y) = \sum\limits_{x}p_{X,Y}(x,y)$
  <br><br>
* With joint PMF, we see that $P(X = Y)$ is sum of diagonal probabilities x = y!
* Can have more than 2 variables:
  * $p_{X,Y,Z}(x,y,z) = P(X = x\ and\ Y = y\ and\ Z = z)$
  * $\sum\limits_{x}\sum\limits_{y}\sum\limits_{z}p_{X,Y,Y}(x,y,z) = 1$
  * $p_X(x) = \sum\limits_{y}\sum\limits_{z}p_{X,Y,Y}(x,y,z)$
  * $p_{X,Y}(x,y) = \sum\limits_{z}p_{X,Y,Y}(x,y,z)$
  <br><br>
* Say we have $Z = g(X,Y)$:
  * $PMF = p_Z(z) = P(Z=z) = P(g(x,y) = z) = \sum\limits_{(x,y):g(x,y)=z}p_{X,Y}(x,y)$
  <br><br>
* Expected Value Rule for Joint PMFs: $\mathbf{E[g(X,Y)] = \sum\limits_{x}\sum\limits_{y}g(x,y)p_{X,Y}(x,y)}$
<br><br>


**Linearity of Expectations for Multiple Random Variables:**
* $\mathbf{E[X + Y] = E[X] + E[Y]}$
  * Proven by defining $g(x,y) = x + y$ and using Expected Value Rule for Joint PMFs
  * Extends to $n$ random variables
  <br><br>
* Note that $E[2X+3Y-Z] = 2E[X] + 3E[Y] - E[Z]$
<br><br>


**Expectation of Binomial Random Variable:**
* Recall Binomial random variable with parameters $n$ and $p$:
  * $X$: number of successes ($P(success) = p)$ and $n$ is number of independent trials
  <br><br>
* Expectation: $E[X] = \sum\limits_{k=0}^{n}kp_X(k) = \sum\limits_{k=0}^{n}k\binom{n}{k}p^k(1-p)^{n-k} \to$ hard to calculate!
  * Break $X$ into $X = X_1 + X_2+...+X_n$, where $X_i = 1$ if $i$'th trial is success, $X_i = 0$ otherwise
  * Now, $E[X] = E[X_1]+...+E[X_n]$ by linearity of expecations
  * Each indicator variable $X_i$ is Bernoulli variable with $E[X_i] = p$
  <br><br>
* Therefore, $\mathbf{E[X] = np}$ for Binomial random variable
<br><br>


**Conditional PMFs Based on Other Random Variables (Joint PMFs):** $A = \{Y=y\}$
* Already seen conditioning PMF of random variable $X$ on event A: $p_{X|A}(x|A) = P(X=x|A)$
* What if A is event that some other random variable takes a value? $\to A = \{Y=y\}$
  * $p_{X|Y}(x|y) = P(X=x|Y=y) = \frac{P(X=x\  \cap\  Y=y)}{P(Y=y)}$
  <br><br>
* This leads to definition of conditioning on another random variable: $\mathbf{p_{X|Y}(x|y) = \frac{p_{X|Y}(x,y)}{p_Y(y)}}$, defined for $y$ where $p_Y(y) > 0$
  * Note that $\sum\limits_{x}P_{X|Y}(x|y) = 1$
  * Also leads to $p_{X|Y}(x,y) = p_Y(y)p_{X|Y}(x|y)$ and similarly, $p_{X|Y}(x,y) = p_X(x)p_{Y|X}(y|x)$
  <br><br>
* Similar forms for more than 2 random variables (can condition on one or more random variables)
* Previously:
  * Multiplication Rule: $P(A \cap B \cap C) = P(A)P(B|A)P(C|A \cap B)$
  * Conditional Expecation: $E[X|A] = \sum\limits_{x}xp_{X|A}(x)$
  * Expected Value Rule: $E[g(X)|A] = \sum\limits_{x}g(x)p_{X|A}(x)$
  * Total Probability Theorem: $p_X(x) = P(A_1)p_{X|A_1}(x)+...+P(A_n)p_{X|A_n}(x)$
    * Given that $\Omega$ partitioned into $A_1, A_2,...,A_n$
    <br><br>
  * Total Expectation Theorem: $E[X] = \sum\limits_{i}P(A_i)E[X|A_i]$
    * Given that $\Omega$ partitioned into $A_1, A_2,...,A_n$
    <br><br>
* Conditioning on $A = \{Y=y\}$
  * Multiplication Rule: if $A = \{X=x\}$, $B = \{Y=y\}$, $C = \{Z=z\} \to p_{X,Y,Z}(x,y,z) = p_X(x) p_{Y|X}(y|x)p_{Z|X,Y}(z|x,y)$
  * Conditional Expectation: $E[X|Y=y] = \sum\limits_{x}xp_{X|Y}(x,y)$
  * Expected Value Rule: $E[g(X)|Y=y] = \sum\limits_{x}g(x)p_{X|Y}(x|y)$
  * Total Probability Theorem: $p_X(x) = \sum\limits_{y}p_Y(y)p_{X|Y}(x|y)$
    * Given that $\Omega$ partitioned into $A_1, A_2,...,A_n$, with $Y = \{y_1,...,y_n\}$ and $A_i = \{Y=y_i\}$ 
    <br><br>
  * Total Expectation Theorem: $E[X] = \sum\limits_{y}p_Y(y)E[X|Y=y]$
    * Given that $\Omega$ partitioned into $A_1, A_2,...,A_n$, with $Y = \{y_1,...,y_n\}$ and $A_i = \{Y=y_i\}$ 
    <br><br>


**Independence of Random Variables:**
* Recall independence of 2 events: $P(A \cap B) = P(A)P(B)$ and $P(A|B) = P(A)$
* Independence of a Random Variable and some event A: $P(X = x \cap A) = P(X=x)P(A)$ for all x
  * PMF of X does not change given A: $p_{X|A}(x) = p_X(x)$ for all x, and by symmetry $P(A|X=x) = P(A)$ for all x
  <br><br>
* For independence of Two Random Variables: $P(X = x \cap Y = y) = P(X=x) P(Y=y)$ for all $x, y$
  * Formally, $\mathbf{p_{X,Y}(x,y) = p_X(x)p_Y(y)}$ for all $x,y$ if Random Variables are independent
  * Can say $p_{X|Y}(x|y) = p_X(x)$ and $p_{Y|X}(y|x) = p_Y(y)$
  * Extends to $n$ random variables
  <br><br>


**Independence and Expectation:**
* If $X$ and $Y$ are independent: $\mathbf{E[XY] = E[X]E[Y]}$
* Additionally, if $g(X)$ and $h(Y)$ are also independent: $E[g(X)h(Y)] = E[g(X)]E[h(Y)]$
  * Can prove these with Expected Value Rule for Joint PMFs
  <br><br>


**Independence and Variance:**
* Recall that $var(aX) = a^2var(X)$ and $var(X+a) = var(X)$, and these are always true
* If $X$ and $Y$ independent, $\mathbf{var(X+Y) = var(X) + var(Y)} \to$ NOT true generally
<br><br>


**Variance of Binomial Random Variable:** $X$: number of successes in $n$ independent trials with $P(success) = p$
* Break $X$ into $X = X_1 + X_2+...+X_n$, where $X_i = 1$ if $i$'th trial is success, $X_i = 0$ otherwise
* Now, $var(X) = var(X_1)+...+var(X_n) = n*var(X_i)$ by independence of indicator variables $X_i$
* Each indicator variable $X_i$ is Bernoulli variable with $var(X_i) = p(1=p)$
* As such, $\mathbf{var(X) = np(1-p)}$
<br><br>

## Additional Theoretical Material and Summary:

**Functions:**
* Definition of function: 2 sets, A and B. Function is a rule that for every element in A, there is a corresponding element in B.
  * Multiple elements of A can map to the same element in B
  * Each element in A only maps to a single element in B
  <br><br>
* Formally, set of pairs $(x,y)$ with $x \in A$ and $y \in B$ where each $x$ appears in exactly one pair
  * Example for square function: $\{(x, x^2) \in \mathbb{R}^2 | x \in \mathbb{R}\}$
  * Equivalently, function $f: \mathbb{R} \to \mathbb{R}$ defined by $f(x) = x^2$, informally "the function $x^2$"
<br><br>


**Variance of Geometric PDF:**
* Let $X$ be a geometric random variable with parameter $p \to$ e.g. number of coin flips that it takes to obtain first heads and $P(heads) = p$
* Recall memorylessness: Given $X>1$, then $X - 1$ has same geometric PMF as unconditional PMF of $X$
  * Can also say that given $X>1$, then $X$ has same geometric PMF as unconditional PMF of $X + 1$ (just added 1)
  <br><br>
* Because of this, $E[X|X>1] = 1 + E[X]$ and $E[X^2|X>1] = E[(X+1)^2]$
* Using Total Expectation Theorem with fact that $E[X] = \frac{1}{p}$ for geometric r.v.s:  
\begin{align}
E[X^2] &= P(X=1)E[X^2|X=1] + P(X>1)E[X^2|X>1] \\
&= p \bullet 1 + (1-p) \bullet (E[X^2] + 2E[X] + 1) \\
&= p + (1-p)(E[X^2] + \frac{2}{p} + 1)
\end{align}
  * Solve for $E[X^2] = \frac{2 - p}{p^2}$
  <br><br>
* Finally, use formula $var(X) = E[X^2] - (E[X])^2$ to get $\mathbf{var(X)_{geo} = \frac{1 - p}{p^2}}$
<br><br>


**Inclusion Exclusion Formula:**

```python
display(Image(filename='incex1.png', embed=True, width=800))
```

* Recall first inclusion exclusion formula for 2 events: $P(A \cup B) = P(A) + P(B) - P(A \cap B)$

* Can generalize to 3 events: $P(A \cup B \cup C) = P(A) + P(B) + P(C) - (P(A \cap B) + P(A \cap C) + P(B \cap C)) + P(A \cap B \cap C)$

* Want to generalize to $n$ joint events using Indicator Functions:
  * For any event $A_i$, give indicator function $X_i$ where $X_i = 1$ if event is in set $A_i$ and $0$ otherwise
  * Note that for $A_u^c$, get $1 - X_i$ (opposite results for indicator variable)
  * For $A_i \cap A_j$, indicator variable is $X_i X_j$, so equal to 1 only if event is in both sets
  * For $A_i^c \cap A_j^c$, indicator variable is $(1-X_i)(1-X_j)$
  * For $A_i \cup A_j$, from DeMorgan's laws, know this is $(A_i^c \cap A_j^c)^c \to$ indicator variable is $1 - (1-X_i)(1-X_j)$
  <br><br>

* Can extend these formulas to any number of sets

* Recall that $P(event) = E[indicator\ of\ event]$
\begin{align}
P(A \cup B \cup C) &= E[1 - (1 - A)(1 - B)(1 - C)] \\
&= E[1 - 1 + A + B + C - AB - BC - AC + ABC] \\
&= E[A + B + C - AB - BC - AC + ABC] \\
&= P(A) + P(B) + P(C) - P(A \cap B) - P(B \cap C) - P(A \cap C) + P(A \cap B \cap C) 
\end{align}
  * This matches above formula of generalizing to three events! We extend this idea to any number of subsets, completing the proof.
<br><br>

**#5: Conditioning example:**

```python
display(Image(filename='sp5_1.png', embed=True, width=900))
```

* Here, we condition a geometric distribution on another independent geometric distribution and arrive at a uniform PMF (dependent only on the sum of the outcomes of the distributions). To make this intuitive, imagine two friends flipping coins. The first takes only 1 flip to get heads (X = 1), and the second takes 9 flips (Y = 9). This means n = 10. Because X = 1 is relatively likely and Y = 9 is relatively unlikely, they "cancel" the effect of one another. So as long as the two friends sum to 10, the probability that it took the first friend i flips to get heads is uniform on $i = 1, 2,..., n-1$
<br><br>

* Recall for geometric distribution (number of indpendent trials until success occurs): $p_X(k) = (1 - p)^{k - 1}p$

* Conditioning the expression: $P(X = i | X + Y = n) = \frac{P(X = i) \cap P(X + Y = n)}{P(X + Y = n)}$

* Numerator: note independence of X and Y:  

\begin{align}
P(X = i) \cap P(X + Y = n) &= P(X = i) \cap P(Y = n - i) \\
&= P(X=i) \bullet P(Y = n-i) \\
&= (1 - p)^{i - 1}p \bullet (1 - p)^{n - i - 1}p \\
&= (1 - p)^{n - 2}p^2
\end{align}

* Denominator: use Total Probability Theorem  
\begin{align}
P(X + Y = n) &= \sum\limits_{k = 1}^{n - 1}P(X = k)P(X + Y = n | X = k) \\
&= \sum\limits_{k = 1}^{n - 1}(1 - p)^{k - 1}p P(Y = n - k) \\
&= \sum\limits_{k = 1}^{n - 1}(1 - p)^{k - 1}p (1 - p)^{n - k - 1}p \\
&= \sum\limits_{k = 1}^{n - 1}(1 - p)^{n - 2}p^2
\end{align}
  * Note: removed conditioning because X and Y are independent
  * Because sum does not depend on $k$, just making copies: $\sum\limits_{k = 1}^{n - 1}(1 - p)^{n - 2}p^2 = (n-1)(1 - p)^{n - 2}p^2$
  <br><br>
  
* Finally, $P(X = i | X + Y = n) = \frac{1}{n-1}$ cancelling everything out
<br><br>

**#9:Indicator variables: the number of inversions. There are n persons, numbered 1 to n. Each person i is assigned a seat number Xi. The seat numbers are distinct integers in the range 1,…,n. **For any i and j, with 1 ≤ i < j ≤ n, we say that we have an inversion if Xi>Xj. Let N be the number of inversions. Find E[N].**

* Counting approach: $\sum\limits_{\sigma} \frac{1}{n!} \bullet num\ inversions\ for\ perm\ \sigma \to$ this is hard and computationally intensive!

  * Now, $N = \sum\limits_{i<j}V_{i,j}$
  * Additionally, $E[N] = E[\sum\limits_{i<j}V_{i,j}] = \sum\limits_{i<j}E[V_{i,j}] = \sum\limits_{i<j}P(X_i > X_j)$ using linearity of expectations and definition of indicator variable
    * Tricky part: because uniformly random permutation, $P(X_i > X_j) = P(X_i < X_j) = \frac{1}{2}$
    <br><br>

* Number of pairings where $i \neq j$ is $n(n-1)$. Divide by 2 because for all pairs e.g. {1,3}, there is a pair {3,1}.
* As such, $E[N] = \frac{1}{2} \frac{n(n-1)}{2} = \frac{1}{4}n(n-1)$

**#8: Indicator variables: the problem of joint lives. Consider 2m persons forming m couples who live together at a given time. Suppose that at some later time, the probability of each person being alive is p, independently of other persons. At that later time, let A be the number of persons that are alive and let S be the number of couples in which both partners are alive. For any number of total surviving persons a, find E[S∣A=a].**
* Define 2 indicator variables:
  * $X_i = 1$ if first partner in couple alive, $X_i = 0$ otherwise
  * $Y_i = 1$ if second partner in couple alive, $Y_i = 0$ otherwise
* Define $Z_i = X_iY_i$, such that $Z_i = 1$ if both couples still alive, $Z_i = 0$ otherwise
  * Now, $S = \sum\limits_{i = 1}^{m}Z_i = \sum\limits_{i = 1}^{m}X_iY_i$
* $E[S|A=a] = E[\sum\limits_{i = 1}^{m}X_iY_i|A=a] = \sum\limits_{i = 1}^{m}E[X_iY_i|A=a]$ using linearity of expectations

\begin{align}
E[S|A=a] &= \sum\limits_{i = 1}^{m}E[X_iY_i|A=a] \\
&=m \bullet E[X_1,Y_1|A=a]\ (1\ being\ arbitrary) \\
&=m \bullet P(X_1Y_1 = 1|A=a)\ (Z_1\ is\ indicator\ variable) \\
&=m \bullet P(X_1 = 1|A=a)P(Y_1 = 1|X_1 = 1 \cap A=a)\ (multiplication\ rule) \\
&=m \bullet \frac{a}{2m} \frac{(a-1)}{(2m-1)} \\
&=\frac{a}{2}\frac{(a-1)}{(2m-1)}
\end{align}

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from math import floor

def expectedSurviving(m, a):
    return a * (a - 1) / (2.0 * (2.0 * m - 1.0))

ppl = 100
m = np.arange(1, floor(ppl/2) + 2) # number of couples
a = np.arange(1, 2 * floor(ppl/2) + 2) # number of surviving people

M, A  = np.meshgrid(m, a)
mask = [A >= 2*M]
A[mask] = 0

Z = expectedSurviving(M, A) # average number of surviving couples
plt.figure(figsize = (16, 8))
im = plt.imshow(Z, aspect='auto')
plt.colorbar(im)
plt.title('Average Number of Surviving Couples with ' + str(ppl) + ' People', fontsize=16)
plt.xlabel('Number of Initial Couples', fontsize=14)
plt.ylabel('Number of Surviving People', fontsize=14)
```
